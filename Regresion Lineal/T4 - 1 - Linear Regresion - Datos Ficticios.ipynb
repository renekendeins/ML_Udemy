# Modelos de Regresion Lineal
## Modelo con datos simulados
* y = a + b * x
* X : 100 valores distribuidos segun una normal de media 1.5 y desv. stand. de 2.5 => N(1.5, 2.5)
* Ye = 10 + 1 * x + e
* e estara destribuida segun una N(0,0.8)

```python
import pandas as pd
from IPython.display import Image
import numpy as np
```

```python
x = 1.5 + 2.5 * np.random.randn(100)
```

```python
# Calculamos los errores. res = residuo
res = 0 + 0.8 * np.random.randn(100)
```

```python
yPred = 5 + 1.9 * x
```

```python
yActual = 5 + 1.9 * x + res
```

```python
# Creamos listas con los datos que tenemos

xLIst = x.tolist()
yPredList = yPred.tolist()
yActualList = yActual.tolist()
```

```python
# Creamos un DataFrame con los valores 
data = pd.DataFrame(
    {
        'x': xLIst,
        'Y_Actual': yActualList,
        'Y_Prediccion' : yPredList
    }
)
```

```python
data.head()
```

```python
import matplotlib.pyplot as plt
```

```python
yMean = [np.mean(yActual) for i in range(1, len(xLIst) + 1)]
```

```python
%matplotlib inline
# Los puntos azules son la prediccion, el modelo lineal
plt.plot(x, yPred)

# Valores de la muestra que hemos tomado, los valores actuales
plt.plot(x, yActual, 'ro')

# Representamos el promedio de los valores
plt.plot(x, yMean, 'g')
plt.title('Valore actual vs Prediccion')

plt.savefig('./fig.png', dpi = 600)
```

```python
data['SSR'] = (data['Y_Prediccion']-np.mean(yActual))**2
data['SSD'] = (data['Y_Prediccion'] - data['Y_Actual'])**2
data['SST'] = (data['Y_Actual'] - np.mean(yActual))**2
```

## Obteniendo la recta de regresion
* y = a + b*x **(Recta de regresion lineal)**
* b = sum((xi-x_m)*(yi-y_m))/sum((xi-x_m)***2) **(Cov/Var)**
* a = y_m - b * x_m


```python
xMean = np.mean(data['x'])
yMean = np.mean(data['Y_Actual'])

xMean, yMean
```

```python
data['covarianza'] = (data['x'] - xMean) * (data['Y_Actual'] - yMean)
data['varianza'] = (data['x'] - xMean)**2
```

```python
beta = sum(data['covarianza'])/sum(data['varianza'])
```

```python
alpha = yMean - beta * xMean
```

```python
alpha, beta
```

El modelo lineal obtenido por regresion

Alpha y beta nos indican que en nuestro modelo, 'y' va a ser:

y = 5.022507779632872 + 1.8944008793071478 * x

```python
data['yModel'] = alpha + beta * data['x']
data.head()
```

```python
SSR = sum((data['yModel']- yMean )**2)
SSD = sum((data['yModel'] - data['Y_Actual'])**2)
SST = sum((data['Y_Actual'] - yMean)**2)
```

```python
SSR,SSD,SST
```

```python
R2 = SSR / SST
R2
```

```python
yMean = [np.mean(yActual) for i in range(1, len(xLIst) + 1)]

%matplotlib inline
# Los puntos azules son la prediccion, el modelo lineal
plt.plot(x, yPred)

# Valores de la muestra que hemos tomado, los valores actuales
plt.plot(x, yActual, 'ro')

# Representamos el promedio de los valores
plt.plot(x, yMean, 'g')

plt.plot(data['x'], data['yModel'])

plt.title('Valore actual vs Prediccion')

plt.savefig('./fig2.png', dpi = 600)
```

## Error estandard de los residuos (RSE)

```python
RSE = np.sqrt(SSD/(len(data)-2))
RSE
```

Esta desviacion tipoca de los residuos, cuando menor sea mucho mejor es el modelo. Comparar este valor con el promedio de los valores actuales es la mejor forma

```python
np.mean(data['Y_Actual'])
```

Si el promedio es 8.047..., 0.7279... sobre el primer valor es realmente pequeno. Podemos decir que hay un error de 0.72 sobre 8.047. Que traducido a porcentage, para que nos hagamos una idea :

```python
RSE / np.mean(data['Y_Actual'])
```

Hay un 9% de error aproximadamente. Por tanto, se observa incrementos de un 9% del error debido al uso de este modelo.
