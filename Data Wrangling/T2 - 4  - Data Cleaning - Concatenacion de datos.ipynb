## Concatenar y apendizar data sets
* Concatenar datasets
    > `whineData = pd.concat([redWine, whiteWine], axis = 0)`

```python
import pandas as pd
```

```python
redWine = pd.read_csv('../../datasets/wine/winequality-red.csv', sep = ';')
redWine.head()
```

```python
whiteWine = pd.read_csv('../../datasets/wine/winequality-white.csv', sep = ';')
whiteWine.head()
```

###### En python tenemos dos tipos de ejes
* axis = 0 -> eje horizontal
* axis = 1 -> eje vertical 

```python
# Concatenamos los DataSets en el axis 0 

whineData = pd.concat([redWine, whiteWine], axis = 0)
```

```python
whineData.shape
```

```python
dw1 = whineData.head(10)
dw2 = whineData[300:310]
dw3 = whineData.tail(10)
```

```python
# El resultado sera un dataset con los datos concatenados tal y como los hemos puesto

dw123 = pd.concat([dw1, dw2, dw3])
dw123
```

## Datos distribuidos

```python
data = pd.read_csv('../../datasets/distributed-data/001.csv')
data.head()
```

* Importar el primer fichero
* Hacemos un bucle para recorrer todos los ficheros. 
    * Importante tener una consistencia en el nombre de los ficheros
    * Importamos los ficheros uno a uno
    * Cada uno debe apendizarse al primer fichero que habiamos cargado
* Repetimos el bucle hasta que no queden ficheros

```python
filePath  ='../../datasets/distributed-data/'

data = pd.read_csv('../../datasets/distributed-data/001.csv')

# Contar el numero de ficheros de un directorio
import os, os.path
DIR = '../../datasets/distributed-data/'
files = len([name for name in os.listdir(DIR) if os.path.isfile(os.path.join(DIR, name))])



for i in range(2,files + 1):
    if i < 10:
        fileName = '00' + str(i)
    elif 10 <= i < 100:
        fileName = '0' + str(i)
    else:
        fileName = str(i)
    
    fileName += '.csv'
    file = filePath + fileName
    
    tempData = pd.read_csv(file)

    
    
    data = pd.concat([data, tempData], axis = 0)
    
data
        
        
    
```

```python
data = data.dropna(axis = 0, how='any')
data
```
